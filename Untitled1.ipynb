```python
import matplotlib.pyplot as plt
 
figure,axes = plt.subplots()
circle = plt.Circle((0.6 ,0.6 ),0.2 )
 
axes.set_aspect( 0.5 )
axes.add_artist( circle )
plt.title( 'Colored Circle' )
plt.show()
```
